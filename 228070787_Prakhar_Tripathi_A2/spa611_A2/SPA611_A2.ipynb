1.) Verify that the mean of power is centered around the variance of the voltage

```python
import numpy as np

# Generate signal
N = 100000
v = np.random.normal(0, 2, N)

power = np.mean(v**2)
variance = np.var(v)

print("Mean Power:", power)
print("Variance:", variance)
```

For a zero-mean random signal, the mean power equals the variance. The above code confirms this as both computed values are nearly identical, validating the theoretical relationship.

2.) Consider cases of different RMS and plot their histograms.

```python
import matplotlib.pyplot as plt

N = 100000

sigmas = [1, 2, 4]

for sigma in sigmas:
    v = np.random.normal(0, sigma, N)
    plt.hist(v, bins=100, density=True, alpha=0.5, label=f"RMS={sigma}")

plt.legend()
plt.title("Histograms for Different RMS Values")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.show()
```

Increasing RMS increases the spread of the distribution while maintaining the Gaussian shape. Higher RMS results in a wider histogram with a lower peak.

3.) How will the histogram change if I add a slowly varying sine wave in it.

```python
t = np.linspace(0, 10, N)
noise = np.random.normal(0, 1, N)
sine = 2 * np.sin(0.5 * t)

v = noise + sine

plt.hist(v, bins=100, density=True)
plt.title("Histogram with Slowly Varying Sine Wave")
plt.xlabel("Voltage")
plt.ylabel("Density")
plt.show()
```

Adding a slowly varying sine wave introduces a deterministic component that shifts the mean over time. This results in a broadened and possibly distorted histogram compared to pure Gaussian noise.

4.) How are the Variances of the ensembles distributed ?

```python
num_ensembles = 500
N = 1000

variances = []

for _ in range(num_ensembles):
    v = np.random.normal(0, 2, N)
    variances.append(np.var(v))

plt.hist(variances, bins=50, density=True)
plt.title("Distribution of Variances Across Ensembles")
plt.xlabel("Variance")
plt.ylabel("Density")
plt.show()

print("Mean of estimated variances:", np.mean(variances))
```

The variance computed from each ensemble is itself a random variable. Its distribution is centered around the true variance and becomes tighter as the number of samples increases, consistent with statistical estimation theory.
